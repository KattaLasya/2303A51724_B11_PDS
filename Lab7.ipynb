```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Create sample dataset (30 days of mock temperature data)
data = pd.DataFrame({
    "Day": range(1, 31),  # Days 1 to 30
    "Temperature": [28, 29, 30, 31, 32, 33, 34, 32, 31, 29,
                    28, 27, 26, 27, 28, 30, 31, 32, 34, 35,
                    36, 35, 33, 32, 30, 29, 28, 27, 28, 29]
})

# 2. Plot line chart
plt.figure(figsize=(10, 5))
plt.plot(data["Day"], data["Temperature"], marker="o", linestyle="-", linewidth=2, color="blue", label="Temperature")

# 3. Highlight hottest and coldest days
hottest_idx = data["Temperature"].idxmax()
coldest_idx = data["Temperature"].idxmin()

hottest_day = data.loc[hottest_idx, "Day"]
hottest_temp = data.loc[hottest_idx, "Temperature"]

coldest_day = data.loc[coldest_idx, "Day"]
coldest_temp = data.loc[coldest_idx, "Temperature"]

plt.scatter(hottest_day, hottest_temp, color="red", s=100, zorder=5, label=f"Hottest ({hottest_temp}C)")
plt.text(hottest_day, hottest_temp + 0.5, f"{hottest_temp}C", ha="center", color="red")

plt.scatter(coldest_day, coldest_temp, color="green", s=100, zorder=5, label=f"Coldest ({coldest_temp}C)")
plt.text(coldest_day, coldest_temp - 1, f"{coldest_temp}C", ha="center", color="green")

# 4. Formatting
plt.title("Daily Temperature Changes Over a Month")
plt.xlabel("Day")
plt.ylabel("Temperature (C)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Create sample dataset (Quarterly Sales)
data = pd.DataFrame({
    "Quarter": ["Q1", "Q2", "Q3", "Q4"],
    "Sales": [12000, 18000, 25000, 16000]  # mock sales values
})

# 2. Plot line chart
plt.figure(figsize=(8, 5))
plt.plot(data["Quarter"], data["Sales"], marker="o", linestyle="-", linewidth=2, color="blue", label="Sales")

# 3. Highlight peak sales with a marker
peak_idx = data["Sales"].idxmax()
peak_quarter = data.loc[peak_idx, "Quarter"]
peak_sales = data.loc[peak_idx, "Sales"]

plt.scatter(peak_quarter, peak_sales, color="red", s=100, zorder=5, label=f"Peak Sales ({peak_sales})")
plt.text(peak_quarter, peak_sales + 1000, f"{peak_sales}", ha="center", color="red")

# 4. Chart formatting
plt.title("Quarterly Sales Performance")
plt.xlabel("Quarter")
plt.ylabel("Sales")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```
